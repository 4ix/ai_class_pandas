```python
import pandas as pd
# import seaborn as sns

# mpg = sns.load_dataset(mpg)
```

```python
df = pd.read_csv('./data/auto-mpg.csv', header=None)

#df 컬럼 명 지정 -> df.columns = [컬럼명, ... , ...] -> df의 컬럼명 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']

# 처음 5행의 자료 확인: df.head()
print(df.head())

print()

# 마지막 5행의 자료 확인: df.tail()
print(df.tail())
```

```python
# 1-2. 데이터 요약 정보 확인하기
# 데이터프레임의 크기(행, 열) -> df.shape: (행, 열)
print(df.shape)

# 데이터프레임의 기본 정보 확인: df.info() -> 어디에 있던지 output으로 결과 출력함
df.info()
```

```python
# 데이터프레임의 각 컬럼의 데이터타입 확인: df.dtypes
print(df.dtypes)
print()
# mpg 컬럼의데이터 타입 확인: df.mpg.dtype
print(df.mpg.dtype)
print()

print(df['mpg'].dtype)
print()

print(df[['mpg','horsepower']].dtypes)
```

```python
# 데이터프레임의 기술 통계 정보 요약 -> df.describe()
print(df.describe()) # 숫자 자료만 보여줌
print()
print(df.describe(include='all')) # 모든 자료 보여줌
```

```python
# 1-3. 데이터 개수 확인
# 각 열의 데이터 개수 확인 -> df.count() 
df.count() # 유효한 값의 개수만 계산 
```

```python
# 각 열의 고유값 개수 -> df['열명'].value_counts()
df['origin'].value_counts()
```

```python
# 2. 통계 함수 적용
# 평균값: df.mean(), df['컬럼명'].mean()
print(df.mean(numeric_only=True))
print()
print(df.mpg.mean()) # mpg 컬럼의 평균
print()
print(df[['mpg','weight']].mean())
```

```python
# 중간값: df.median(), df['컬럼명'].median()
print(df.median(numeric_only=True))
print()
print(df.mpg.median()) # mpg 컬럼의 중간값
```

```python
# 최대값: df.max(), df['컬럼명'].max()
print(df.max())
print()
print(df.mpg.max()) # mpg 컬럼의 최대값
```

```python
# 최소값: df.min(), df['컬럼명'].min()
print(df.min())
print()
print(df.mpg.min()) # mpg 컬럼의 최소값
```

```python
# 표준편차: df.std(), df['컬럼명'].std()
print(df.std(numeric_only=True))
print()
print(df.mpg.std()) # mpg 컬럼의 표준편차
```

```python
# 상관계수: df.corr(), df[['컬럼명','컬럼명']].corr()

print(df.corr())
print(df[['mpg','weight']].corr())
```

```python
# 3. 판다스 내장 그래프 도구 -> df.plot()
# 1. 남북한 발전 전력량 읽어와서 df에 저장
# 2. 0행(남한)과 5행(북한), 3열부터 끝(1991년 이후 자료만)까지의 자료만 추출해서 df_ns에 저장 (iloc)


df = pd.read_excel('./data/남북한발전전력량.xlsx')
df.info()
```

```python
df_ns = df.iloc[[0,5],3:]
df_ns.info()

# 인덱스명을 수정 -> 남한, 북한으로
df_ns.index = ['South','North']
df_ns
```

```python
df_ns.plot()
```

```python
# 전치 행렬로 변환
tdf_ns = df_ns.T # 연도와 인덱스를 바꿈 -> 연도가 인덱스가 되도록
tdf_ns.plot()
```

```python
tdf_ns.plot(kind='bar')
```

```python
tdf_ns.plot(kind='barh')
```

```python
df_ns.columns #컬럼의 값이 문자이므로 숫자로 반환해야 hist 그릴 수 있음
df_ns.columns = df_ns.columns.map(int) # map 함수 이용하여 int로 변경함
df_ns.columns
```

```python
tdf_ns = df_ns.T
tdf_ns.info()
# tdf_ns.plot(kind='hist') # x축(컬럼)이 숫자 범위여야 함
```
