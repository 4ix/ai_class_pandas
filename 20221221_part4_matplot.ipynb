```python
# %load matplot.py
# %%
import matplotlib.pyplot as plt
plt.rc('font', family = 'gulim') # For windows, 한글 폰트
plt.rc('axes', unicode_minus=False) # minus 기호 출력
```

```python
import pandas as pd

df = pd.read_excel('./data/시도별 전출입 인구수.xlsx')
df.head()
```

```python
# 1. 누락 데이터 처리: df.fillna(value or method=ffill / bfill)
df1 = df.fillna(0) # NaN에 0를 대체
# df1.head()

# NaN에 엑셀 파일의 특징을 접목해서 앞에 나오는 자료로 대체
df = df.fillna(method='ffill')
# print(df.head())

df.iloc[20:35, :]
```

```python
# 서울에서 다른 지역으로 전출한 이동 인구수의 분포도
# 2. 전출지별이 서울특별시이고 전입지별이 서울특별시가 아닌것
df_서울 = df[(df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')]
df_서울.head()
```

```python
# 3. 불필요한 컬럼 또는 인덱스(행) 삭제: df.drop('컬럼명', axis=1), df.drop('인덱스명', axis=0)
df_서울 = df_서울.drop('전출지별', axis=1)
df_서울.head()
```

```python
# 4. 컬럼명 재정의: '전입지별' -> '전입지'로 변경
# df_서울 = df_서울.rename(columns={'전입지별':'전입지'})
df_서울 = df_서울.rename({'전입지별':'전입지'}, axis=1)
df_서울.head()
```

```python
# 5. 임의의 컬럼으로 인덱스를 설정: '전입지' -> 인덱스로 설정
df_서울 = df_서울.set_index('전입지')
df_서울.head()
```

```python
# 최종: 서울에서 경기도로 전입한 인구의 이동 데이터 추출
df_1 = df_서울.loc['경기도']
df_1
```

```python
# 그래프는 한 셀에서 다 그려줘야 함
plt.figure(figsize=(14,5))
plt.plot(df_1.index, df_1.values, marker ='o', markersize=5, color='red')
plt.title('서울 -> 경기도 인구 이동', size=15) # 그래프의 제목 추가
plt.xlabel('기간', size=12) # 그래프의 x축 label 추가
plt.ylabel('이동 인구수', size=12) # 그래프의 y축 label 추가
plt.xticks(rotation=75)
plt.legend(labels=['서울 - > 경기'])

# y축 범위 지정 (최소값, 최대값)
plt.ylim(50000, 800000)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 620000),       #화살표의 머리 부분(끝점)
             xytext=(2, 290000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
             )

# 주석 표시 - 텍스트
plt.annotate('인구이동 증가(1970-1995)',  #텍스트 입력
             xy=(10, 400000),            #텍스트 위치 기준점
             rotation=25,                #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=15,                #텍스트 크기
             )

plt.annotate('인구이동 감소(1995-2017)',  #텍스트 입력
             xy=(40, 500000),            #텍스트 위치 기준점
             rotation=-9,               #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=15,                #텍스트 크기
            )

plt.show()
```

```python
# 남북한발전전략.xlsx 파일을 읽어서 df에 저장 후 
# NaN 데이터는 ffill로 대체
# 남한과 북한의 합계 자료만 추출
# '전력량 (억kwh)' 컬럼명을 '구분' 으로 , '발전 전력별' 은 삭제
# '구분'을 인덱스로
# 그래프 그리기
# title: 남한과 북한의 연도별 전력 사용량
# xlable: 기간
# ylabel: 사용량
```

```python
df = pd.read_excel('./data/남북한발전전력량.xlsx')
df = df.fillna(method='ffill')
df = df[(df['발전 전력별']=='합계')]
df = df.rename({'전력량 (억㎾h)':'구분'}, axis=1)
df = df.drop('발전 전력별', axis=1)
df = df.set_index('구분')
# df.'남한' = df.남한.map(int)
# df_I.북한 = df.북한.map(int)
# df.info()


df = df.T # 연도와 인덱스를 바꿈 -> 연도가 인덱스가 되도록
print(df.head())

plt.figure(figsize=(14,5)) # 바탕 사이즈
plt.plot(df.index, df.values, marker="*")
plt.title('남한과 북한의 연도별 전력 사용량')
plt.xlabel('기간')
plt.ylabel('사용량')
plt.xticks(rotation='vertical')

plt.legend(labels=['남한','북한'], loc='best') # loc = 'center', 'best', .. ## 범례 표시
plt.show()
```

```python
# 서울 -> 충남, 충북, 전북, 전남으로 이동한 인구수를 각각의 그래프로 표현
# 연도는 1990년 부터 2017년 까지 -> int형으로 변경
# title, legend, xtick은 75도 회전
# xlabel은 기간, ylabel은 이동 인구수


df_충청남도 = df_서울.iloc[11, 20:]
df_충청북도 = df_서울.iloc[10, 20:]
df_전라북도 = df_서울.iloc[12, 20:]
df_전라남도 = df_서울.iloc[13, 20:]

df_충청남도 = df_충청남도.astype(int)
df_충청북도 = df_충청북도.astype(int)
df_전라북도 = df_전라북도.astype(int)
df_전라남도 = df_전라남도.astype(int)

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,1,1)

ax.plot(df_충청남도, marker = 'o', label = '서울 -> 충청남도')
ax.plot(df_충청북도, marker = '*', markerfacecolor='green', color='olive', label = '서울 -> 충청북도')
ax.plot(df_전라북도, marker = '^', markerfacecolor='black', color='yellow', label = '서울 -> 전라북도')
ax.plot(df_전라남도, marker = '^', markerfacecolor='red', color='blue', label = '서울 -> 전라남도')

ax.legend()
ax.set_title('서울 -> 충남, 충북, 전북, 전남 이동 인구수', size=15)
ax.set_xlabel('기간')
ax.set_ylabel('이동 인구수')

ax.tick_params(axis='x', labelrotation=75)

plt.show()
```

```python
# 2축 그래프 그리기
# 남북한 전력 발전량 수력과 화력(막대 그래프), 발전량 증감
# 남북한발전전력량 파일 읽어서 df에 저장
# 북한의 자료만 추출

df = pd.read_excel('./data/남북한발전전력량.xlsx')
df = df.iloc[5:]
df = df.drop('전력량 (억㎾h)', axis=1)
df = df.set_index('발전 전력별')
```

```python
# df = df.rename({'합계':'총발전량'}, axis=1)
# df
df1 = df.T
df1
```

```python
# 이전 년도 총 발전량을 컬럼에 추가: "이전년도 총발전량"
df1['이전년도 총발전량'] = df1.loc[:,'합계'].shift(1)
df1
```

```python
# 중감율 컬럼 추가 -> ((총발전량 / 이전년도 발전량)-1)*100
df1['증감율'] = ((df1['합계'] / df1['이전년도 총발전량'])-1)*100
df1
```

```python
df1.head()
```

```python
ax1 = df1[['수력','화력']].plot(kind='bar', width=0.7, stacked=True, figsize=(12,5))
ax2 = ax1.twinx()
ax2.plot(df1.index,df1.증감율, ls='--', marker='o', color = 'black', label='전년대비 증감율(%)')
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)
ax1.set_xlabel('연도')
ax1.set_ylabel('전력량 (억㎾h)')
ax2.set_ylabel('전년대비 증감율(%)')
plt.title('북한 전력 발전량')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
