```python
import pandas as pd
```

```python
# 시리즈: pd.Series(리스트 또는 딕셔너리)
dict_a={'a':1, 'b':2, 'c':3}

print(dict_a, type(dict_a))

# 딕셔너리로 시리즈 생성
sr = pd.Series(dict_a)
print(sr, '\n', type(sr))
```

```python
# 자료를 접근하는 방법 -> 인덱스로 접근: 키 인덱스 또는 숫자 인덱스
print(sr['a'], sr[0])
print(sr.index) # 시리즈의 인덱스
print(sr.values) # 시리즈의 밸류스
print()
print(sr['a':'c']) # 키 인덱스는 start:stop -> start포함, stop도 포함
print()
print(sr[0:2]) # 숫자 인덱스는 start:stop -> start포함, stop도 포함하지 않음
```

```python
# 4. 산술연산
# 4-1. 시리즈 연산
# 시리즈 연산자 숫자: 시리즈 연산자(+ - * /) 숫자 -> 시리즈의 각 요소에 연산자 적용
student1 = pd.Series({'국어':100, '영어':90, '수학':80})
print(student1)

print(student1 + 100) # 시리즈의 각 데이터에 100을 더한 값을 시리즈로 반환
print(student1 - 100)
print(student1 * 100)
print(student1 / 100)
```

```python
# 시리즈 연산자(+,-,*,/) 시리즈
student2 = pd.Series({'수학':90, '국어':90, '영어':80})
print(student1) # 국어 영어 수학
print(student2) # 수학 국어 영어

# 학생의 과목별 합계를 구함
hap = student1+student2
cha = student1-student2
mul = student1*student2
div = student1/student2

print(hap)

result = pd.DataFrame([hap,cha,mul,div], index=['+','-','*','/'])
print(result)
```

```python
import numpy as np
student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':90, '국어':90})
```

```python
print(student1)
print(student2)

hap = student1+student2
cha = student1-student2
mul = student1*student2
div = student1/student2

# 연산식에 NaN이 포함되면 결과는 NaN
result = pd.DataFrame([hap,cha,mul,div], index=['+','-','*','/'])
print(result)
```

```python
# 연산 메소드 -> sr1.add(sr2)
# add 더하기, sub 빼기, mul 곱하기, div 나누기
hap = student1.add(student2, fill_value=0) # fill_value 입력시 NaN을 해당 값으로 처리함
cha = student1.sub(student2, fill_value=0) # fill_value 입력시 NaN을 해당 값으로 처리함
mul = student1.mul(student2, fill_value=0) # fill_value 입력시 NaN을 해당 값으로 처리함
div = student1.div(student2, fill_value=0) # fill_value 입력시 NaN을 해당 값으로 처리함

result = pd.DataFrame([hap,cha,mul,div], index=['+','-','*','/'])
print(result)
```
